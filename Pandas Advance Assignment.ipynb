### Q1.

#### Five functions of Pandas library are:
1. pd.read_csv: This is used to retrieve data from csv files in the form of a dataframe (say df).
2. df.head(n): It returns the top n (5 by default) values of df or a series.
3. df.sample()L This is used to generate a random sample from the dataframe df.
4. df.describe(): This returns the general statistics, like mean, quantiles, count etc. of all the columns of df in the form of a dataframe.
5. df.dtypes(): This method returns a Series with the data type of each column.

##### Q2. A scatter plot is a diagram which represents values corresponding to x-axis and y-axis using dots.

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
np.random.seed(3)
```

```python
x=3+np.random.normal(0,2,50)
```

```python
y=3+np.random.normal(0,2,len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Scatter Plot of X vs Y")
plt.show()
```

##### Q3. With the subplot() function, users can easily create multiple plots in a single figure, making comparing and analyzing different datasets or variables easier.

```python
#plot 1
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)

#plot2
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)

#plot 3
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)

#plot 4
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
```

##### Q4. Bar plots are a type of data visualization used to represent data in the form of rectangular bars. The height of each bar represents the value of a data point, and the width of each bar represents the category of the data. 
##### Bar graphs are used to compare things between different groups or to track changes over time. It shows the relationship between a numeric and a categoric variable.

```python
company = np.array(['Apple','Microsoft','Google','AMD'])
```

```python
profit = np.array([3000,8000,1000,10000])
```

```python
plt.bar(company,profit)
plt.xlabel("Companies")
plt.ylabel("Profits")
plt.title("Profits of Companies")
```

```python
#Horizontal Bar
plt.barh(company,profit)
plt.xlabel("Companies")
plt.ylabel("Profits")
plt.title("Profits of Companies")
```

##### Q5. Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups. They are built to provide high-level information at a glance, offering general information about a group of data's symmetry, skew, variance, and outliers. It organizes large amounts of data, and visualizes outlier values.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
#boxplot 1
plt.subplot(2,1,1)
plt.boxplot(box1,vert=False)

#boxplot 2 
plt.subplot(2,1,2)
plt.boxplot(box2,vert=False)
plt.show()
```

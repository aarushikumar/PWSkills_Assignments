### Question 3

```python
import numpy as np
```

```python
np.random.seed(0)
```

```python
exercise_hours = np.random.uniform(0, 10, 50)
```

```python
bmi = 30 - (exercise_hours * 0.5) + np.random.normal(0, 2, 50)
```

```python
df = pd.DataFrame({
    'Hours_of_Exercise_per_Week': exercise_hours,
    'BMI': bmi
})
df
```

```python
df.corr(method='pearson')
```

```python
df.corr(method='spearman')
```

In case of both the methods - pearson and spearman, the results are quite similar. The correlation between 'hours of exercise per week' and 'BMI' is moderate weak correlation.

### Question 4.

```python
# Set seed for reproducibility
np.random.seed(42)

# Generate random data for the number of hours spent watching television per day (between 0 and 10)
tv_hours = np.random.uniform(0, 10, 50)

# Generate random data for the level of physical activity with a mild negative correlation with TV hours
# Level of physical activity can be a score between 0 and 100, decreasing slightly with more TV hours
physical_activity = 100 - (tv_hours * 5) + np.random.normal(0, 10, 50)

# Ensure physical activity scores are within the range of 0 to 100
physical_activity = np.clip(physical_activity, 0, 100)

# Create a DataFrame
df_tv = pd.DataFrame({
    'Hours_of_TV_per_Day': tv_hours,
    'Physical_Activity_Level': physical_activity
})

df_tv
```

```python
df_tv.corr(method='pearson')
```

### Question 6.

```python
# Set seed for reproducibility
np.random.seed(42)

# Generate random data for the number of sales calls made per day (between 10 and 50)
sales_calls_per_day = np.random.uniform(10, 50, 30)

# Generate random data for the number of sales made per week with a mild positive correlation with sales calls per day
# Assume a basic linear relationship with some noise
sales_per_week = (sales_calls_per_day * 1.5) + np.random.normal(0, 5, 30)

# Create a DataFrame
df_sales = pd.DataFrame({
    'Sales_Calls_per_Day': sales_calls_per_day,
    'Sales_per_Week': sales_per_week
})

df_sales
```

```python
df_sales.corr()
```

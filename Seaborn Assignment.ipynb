##### Q1. Five plots from the seaborn library are:
1. Scatter Plot: Scatter plots are used to visualize the relationship between two continuous variables. Each point on the plot represents a single data point, and the position of the point on the x and y-axis represents the values of the two variables. sns.scatterplot() 
2. Bar plots: are used to visualize the relationship between a categorical variable and a continuous variable. In a bar plot, each bar represents the mean or median (or any aggregation) of the continuous variable for each category. sns.barplot()
3. Histograms: They visualize the distribution of a continuous variable. In a histogram, the data is divided into bins and the height of each bin represents the frequency or count of data points within that bin. sns.histplot()
4. Density plots (or kernel density plots) are a type of data visualization that display the distribution of a continuous variable. They are similar to histograms, but instead of representing the data as bars, density plots use a smooth curve to estimate the density of the data. sns.kdeplot() 
5. Box plots are a type of visualization that shows the distribution of a dataset. They are commonly used to compare the distribution of one or more variables across different categories. sns.boxplot()

```python
import matplotlib.pyplot as plt
```

```python

sns.lineplot(x='timepoint',y='signal',data=data,hue='event')
plt.title("Lineplot of Signal vs Timepoint based on Event ")
```

```python
sns.lineplot(x='timepoint',y='signal',data=data,hue='region')
plt.title("Lineplot of Signal vs Timepoint based on Region ")
```

```python
sns.lineplot(x='timepoint',y='signal',data=data,hue='region',style='event')
plt.title("Lineplot of Signal vs Timepoint based on Event and Region ")
```

**Question 5**

A Distplot displays a combination of statistical representations of numerical data, such as histogram, kernel density estimation or normal curve, and rug plot. It depicts the variation in the data distribution.



```python
import numpy as np
```

```python
import plotly.figure_factory as ff
```

```python
x = np.random.randn(1000)
```

```python
hist_data = [x]
#hist_data
```

```python
fig = ff.create_distplot(hist_data,['Distplot'])
fig.show()
```
